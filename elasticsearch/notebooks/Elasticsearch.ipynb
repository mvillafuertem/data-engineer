```python
from pprint import pprint as pp
from elasticsearch import Elasticsearch

client = Elasticsearch(['http://elasticsearch:9200'])
pp(client.info())
```

 ## FindById



```python

index = "kibana_sample_data_ecommerce"
id = "L2JlRnEBadSzWcYvqP3r"
findById = client.get(index=index, id=id)
pp(findById)
```

```python
queryByField = client.search(index=index, body={"query": {"match": {'customer_gender':'MALE', "max_query_terms": 1}}})
pp(queryByField)
```
